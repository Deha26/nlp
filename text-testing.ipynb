```python
file=open("Desktop\\newsample.txt","r")
res=file.read()
print("Word count:",len(res))
print(" ".join(res.split()))
```

```python
import re
def decontracted(phrase):
    phrase=re.sub(r"n\'t"," not",phrase)
    phrase=re.sub(r"\'re"," are",phrase)
    phrase=re.sub(r"\'s"," is",phrase)
    phrase=re.sub(r"\'ll"," will",phrase)
    phrase=re.sub(r"\'t"," not",phrase)
    phrase=re.sub(r"\'ve"," have",phrase)
    phrase=re.sub(r"\'m"," am",phrase)
    return phrase
res=decontracted(res)
print(decontracted(res))
```

```python
import nltk,re
from nltk.tokenize import word_tokenize
new=re.sub(r'[^\w\s]','',res)
tokens=word_tokenize(new)
print("Unique words:",len(tokens))
print(tokens)
```

```python
from collections import Counter
freq1=Counter(new.split()).most_common()
print("Words with frequency:",freq1)
```

```python
dic={}
ci={}
for i in range(len(freq1)):
    also=round(freq1[i][1]/len(res),5)
    dic[freq1[i][0]]=also
    c=round(freq1[i][1],5)
    ci[freq1[i][0]]=c
```

```python
pairs=[" ".join(pair) for pair in nltk.bigrams(tokens)]
print(pairs)
```

```python
pairs1=[" ".join(pair) for pair in nltk.trigrams(tokens)]
print(pairs1)
```

```python
freq2=Counter(pairs).most_common()
print("No. of pairs:",len(freq2))
print("Pairs with freq:",freq2)
```

```python
freq3=Counter(pairs1).most_common()
print("No. of pairs:",len(freq3))
print("Pairs with freq:",freq3)
```

```python
#Hypothesis Testing
```

```python
import math 
for i in range(0,3):
    u=round(freq2[i][1]/len(res),5)
    print(freq2[i][0],"--> expexted mean:",u)
    p=round(freq2[i][1]/len(res),5)
    s=freq2[i][0].split(" ")
    ss=p*(1-p)
    num=u-dic[s[0]]*dic[s[1]]
    den=math.sqrt(ss/len(res))
    final=num/den
    print(final)
    critic=2.576
    if final>critic:
        print("Reject null hypothesis and accept",freq2[i],"is a collocation")
    else:
        print("Not a collocation")
```

```python
#Chi-square test
```

```python
for i in range(0, 3):
    s=freq2[i][0].split(" ")
    cw1 = ci[s[0]]
    cw2 = ci[s[1]]
    cnw1 = len(res) - cw1
    cnw2 = len(res) - cw2
    u = round(freq2[i][1] / len(res), 5)
    ow1w2 = u
    onw1w2 = cw1 - ow1w2
    ow1nw2 = cw2 - ow1w2
    onw1nw2 = len(res) - u
    #print("Observed:")
    #print(ow1w2, " ", onw1w2, " ", onw1nw2, " ", ow1nw2)
    ew1w2 = (cw1 * cw2) / len(res)
    enw1w2 = (cnw1 * cw2) / len(res)
    ew1nw2 = (cnw2 * cw1) / len(res)
    enw1nw2 = (cnw2 * cnw1) / len(res)
    #print("Expected:")
    #print(ew1w2, " ", enw1w2, " ", enw1nw2, " ", ew1nw2)
    xw1w2 = pow((ow1w2 - ew1w2), 2) / ew1w2
    xnw1w2 = pow((onw1w2 - enw1w2), 2) / enw1w2
    xw1nw2 = pow((ow1nw2 - ew1nw2), 2) / ew1nw2
    xnw1nw2 = pow((onw1nw2 - enw1nw2), 2) / enw1nw2
    #print("Chi-Square:")
    #print(xw1w2, " ", xnw1w2, " ", xnw1nw2, " ", xw1nw2)
    chi_value = xw1w2 + xnw1w2 + xw1nw2 + xnw1nw2
    chi_constant = 3.84
    print("Chi-square for '", freq2[i][0], "':", str(chi_value))
    if chi_value > chi_constant:
        print("Reject null hypothesis and accept", freq2[i][0], "is a collocation.")
    else:
        print("Not a collocation")
```
